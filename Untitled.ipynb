```python
def get_fm_cnt(cfg):
    fm_cnt=0
    for l in cfg:
        if l != 'M':
            fm_cnt += l[0]
    return fm_cnt

def get_flops(cfg,img_h,img_w):
    flops = 0
    in_channles=3
    for l in cfg:
        if l != 'M':
            flops += in_channles*l[0]*l[1]*l[1]*img_h*img_w
            in_channles = l[0]
        else:
            img_h=img_h/2
            img_w=img_w/2
    return flops
```
